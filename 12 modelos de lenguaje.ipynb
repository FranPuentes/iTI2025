```python
# prompt: Carga un modelo de HF para calcular embeddings de textos.

!pip -q install transformers

from transformers import AutoTokenizer, AutoModel
import torch

# Reemplaza con el nombre del modelo que deseas usar
model_name = "sentence-transformers/all-MiniLM-L6-v2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# prompt: Usando el modelo cargado, calcula los embeddings de cada índice de 'responses'.

# Calcula los embeddings de cada índice de 'responses'
response_embeddings = []
for response in responses.values():
  inputs = tokenizer(response, return_tensors="pt")
  with torch.no_grad():
    embedding = model(**inputs).pooler_output
  response_embeddings.append(embedding)
```

```python
# prompt: Crea un bucle que pregunte al usuario por un texto, calcule su embedding y busque el embedding de 'responses' más próximo, imprimiendo el valor que tienen en 'responses'. El buble finaliza cuando el usuario introduce un texto vacío.

while True:
  user_input = input("Introduce un texto (o pulsa Enter para salir): ")
  if user_input == "":
    break

  # Calcula el embedding del texto introducido por el usuario
  inputs = tokenizer(user_input, return_tensors="pt")
  with torch.no_grad():
    user_embedding = model(**inputs).pooler_output

  # Encuentra el embedding de 'responses' más próximo
  best_similarity = float('-inf')
  best_response = None
  for i, response_embedding in enumerate(response_embeddings):
    similarity = torch.cosine_similarity(user_embedding, response_embedding, dim=1).item()
    if similarity > best_similarity:
      best_similarity = similarity
      best_response = list(responses.keys())[i]

  # Imprime la respuesta correspondiente
  print(responses[best_response])
```
