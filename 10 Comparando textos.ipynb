## *Bag of words*

La "bolsa de palabras" es una técnica simple que indentifica un texto. Consiste en el conjunto de las palabras que hay en dicho texto.

```python
from collections import Counter;
import re;

def bag_of_words(text):
    # Convierte el texto a minúsculas
    text = text.lower();
    # Usa expresiones regulares para reemplazar todo lo que no sea letra, dígito o espacio con un espacio
    text = re.sub(r'[^a-z0-9áéíóúäëïöüñ ]', ' ', text);
    # Divide el texto en palabras
    words = text.split();
    # Crea un contador para las palabras
    return Counter(words);
```

```python
# prompt: Dado texto1, text2 y text3 calcula la BOW de los tres textos y muestra sus palabras más frecuentes y las menos frecuentes.

# Calcula la BOW de los tres textos
bow1 = bag_of_words(texto1)
bow2 = bag_of_words(texto2)
bow3 = bag_of_words(texto3)

# Muestra las palabras más frecuentes
print("Palabras más frecuentes en texto1:", bow1.most_common(5))
print("Palabras más frecuentes en texto2:", bow2.most_common(5))
print("Palabras más frecuentes en texto3:", bow3.most_common(5))

# Muestra las palabras menos frecuentes
print("\nPalabras menos frecuentes en texto1:", bow1.most_common()[:-6:-1])
print("Palabras menos frecuentes en texto2:", bow2.most_common()[:-6:-1])
print("Palabras menos frecuentes en texto3:", bow3.most_common()[:-6:-1])
```

```python
# prompt: Calcula la intersección del BAG de text1 y text2, y de text1 y text3. Antes elimina las 25 palabas más frecuentes del corpus.

# Une los tres textos para crear el corpus
corpus = texto1 + texto2 + texto3

# Calcula la BOW del corpus
bow_corpus = bag_of_words(corpus)

# Obtén las 25 palabras más frecuentes del corpus
stop_words = [word for word, count in bow_corpus.most_common(25)]

# Elimina las stop words de las BOW de los textos
bow1_filtered = {word: count for word, count in bow1.items() if word not in stop_words}
bow2_filtered = {word: count for word, count in bow2.items() if word not in stop_words}
bow3_filtered = {word: count for word, count in bow3.items() if word not in stop_words}

# Calcula la intersección de las BOW filtradas
intersection12 = set(bow1_filtered.keys()) & set(bow2_filtered.keys())
intersection13 = set(bow1_filtered.keys()) & set(bow3_filtered.keys())

# Muestra las intersecciones
print("Intersección de texto1 y texto2 (sin stop words):", intersection12)
print("Intersección de texto1 y texto3 (sin stop words):", intersection13)
```

**La primera aproximación que podemos usar es comparar la *bag of words* de ambos documentos, a mayor coincidencia de palabras, mayor probabilidad de que ambos documentos traten el mismo tema**.

## **Distancia Jaccard**

Esta métrica mide la similitud entre conjuntos y es útil para comparar documentos en términos de las palabras que contienen. Se define como el tamaño de la intersección de dos conjuntos dividido por el tamaño de su unión. Es especialmente útil cuando solo te interesan las palabras presentes sin considerar su frecuencia.

```python
# prompt: Calcula la distancia jaccard entre los tres textos, excepto entre ellos mismos. Elimina previamente las 25 stop words del corpus.

def jaccard_distance(set1, set2):
  intersection = set1 & set2
  union = set1 | set2
  return 1 - (len(intersection) / len(union))

# Calcula la distancia Jaccard entre los textos (sin stop words)
distance12 = jaccard_distance(set(bow1_filtered.keys()), set(bow2_filtered.keys()))
distance13 = jaccard_distance(set(bow1_filtered.keys()), set(bow3_filtered.keys()))
distance23 = jaccard_distance(set(bow2_filtered.keys()), set(bow3_filtered.keys()))

print("Distancia Jaccard entre texto1 y texto2 (sin stop words):", distance12)
print("Distancia Jaccard entre texto1 y texto3 (sin stop words):", distance13)
print("Distancia Jaccard entre texto2 y texto3 (sin stop words):", distance23)
```

```python
# Los resultados de la distancia Jaccard indican lo siguiente:

# - Distancia Jaccard entre texto1 y texto2 (sin stop words): 0.8163265306122449
# - Distancia Jaccard entre texto1 y texto3 (sin stop words): 0.95
# - Distancia Jaccard entre texto2 y texto3 (sin stop words): 0.9444444444444444

# La distancia Jaccard varía entre 0 y 1, donde 0 indica que los conjuntos son idénticos y 1 indica que no tienen elementos en común.

# En este caso, la distancia entre el texto1 y el texto2 es menor que la distancia entre el texto1 y el texto3, y entre el texto2 y el texto3.
# Esto sugiere que los textos 1 y 2 son más similares entre sí en términos de las palabras que contienen, en comparación con los otros pares de textos.

# El hecho de que las distancias entre los textos 1 y 3, y 2 y 3 sean cercanas a 1 indica que estos pares de textos tienen muy pocas palabras en común, lo cual es esperable dado que tratan temas diferentes (el hundimiento del Titanic y el montañismo).
```

Otra aproximación es el método Dice, en donde la operación es dos veces la intersección de ambos conjuntos, dividido la suma de los tamaños de ambos conjuntos.

El resultado es similar, en términos relativos, al de Jaccard, pero "estira" más los valores de simulitud comparados.

```python
# prompt: Calcula el coeficiente Dice entre los tres textos, excepto entre ellos mismos. Elimina previamente las 25 stop words del corpus.

def dice_coefficient(set1, set2):
  intersection = set1 & set2
  return 2 * len(intersection) / (len(set1) + len(set2))

# Calcula la distancia Dice entre los textos (sin stop words)
dice12 = dice_coefficient(set(bow1_filtered.keys()), set(bow2_filtered.keys()))
dice13 = dice_coefficient(set(bow1_filtered.keys()), set(bow3_filtered.keys()))
dice23 = dice_coefficient(set(bow2_filtered.keys()), set(bow3_filtered.keys()))

print("Distancia Dice entre texto1 y texto2 (sin stop words):", dice12)
print("Distancia Dice entre texto1 y texto3 (sin stop words):", dice13)
print("Distancia Dice entre texto2 y texto3 (sin stop words):", dice23)
```

Una vez entrenado, dicho modelo puede detectar palabras sinónimas o sugerir palabras adicionales para una frase sin terminar.

## El concepto de atención

El concepto de atención, por medio de una arquitectura de red neuronal artificial denominada *transformer*, nos permite entrenar modelos que tengan en cuenta las **relaciones sintácticas** entre las palabras.

Gracias a estos modelos, podemos obtener fácilmente vectores de características de textos que, aplicados a estos nos entregen un medio para compararlos fácilmente.
