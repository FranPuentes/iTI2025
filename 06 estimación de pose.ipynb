## Estimación de pose (GPU)

```python
# prompt: Dada una imagen referenciada mediante una url, muestra la imagen y lleva a cabo una estimación de pose usando yolo v8.

!pip -q install ultralytics

from ultralytics import YOLO
from PIL import Image
import requests
from io import BytesIO

def estimate_pose(image_url):
  response = requests.get(image_url)
  image = Image.open(BytesIO(response.content))
  display(image)

  model = YOLO('yolov8n-pose.pt')
  results = model(image)
  display(Image.fromarray(results[0].plot()))

# Ejemplo de uso
image_url = 'https://cdn.stocksnap.io/img-thumbs/960w/wheelchair-person_WLAWAIXW2T.jpg'
estimate_pose(image_url)
```
