## Detección de caras (GPU)

```python
# prompt: Dada una imagen referenciada mediante una url, lleva a cabo la detección de caras mostrando la imagen original y en otra imagen las caras detectadas. Usa un modelo de google (mediapipe).

import numpy as np
!pip -q install mediapipe

import cv2
import mediapipe as mp
import urllib.request

from google.colab.patches import cv2_imshow

mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

# Replace with the URL of the image you want to process
image_url = 'https://freerangestock.com/sample/34798/two-men-faces-and-white-powder-.jpg'

with urllib.request.urlopen(image_url) as response:
  image_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

with mp_face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.50) as face_detection:
  # Convert the BGR image to RGB and process it with MediaPipe Face Detection.
  results = face_detection.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

  # Draw face detections of each face.
  if not results.detections:
    print("No faces detected.")
  else:
    annotated_image = image.copy()
    for detection in results.detections:
      mp_drawing.draw_detection(annotated_image, detection)

    # Display the original and annotated images
    cv2_imshow(image)
    cv2_imshow(annotated_image)
```
